```python
import pandas as pd
import requests
import re
from langchain_text_splitters import RecursiveCharacterTextSplitter

def fetch_precedent_content(url : str):
    try :
        response = requests.get(url)
        response.raise_for_status()
        json = response.json()
        if not isinstance(json , dict) or not 'PrecService' in json.keys():
            print(f"{url} : the json file can't be found") # Json 파일이 제공되지 않는 경우
            return
        return response.json().get('PrecService', "")
        
    except requests.exceptions.RequestException as e:
        print(f"Error fetching URL {url} : {e}")
        return
    except Exception as e:
        print(f"An unexpected error occurred while fetching the URL ({url}): {e}")
        return

def precedent_doc_trim_content(content: dict[str, str]):
    return {
        "판시사항": content.get("판시사항", '').strip(),
        "판결요지": content.get("판결요지", '').strip(),
        "사건명": content.get("사건명", '').strip(),
        "판례내용": content.get("판례내용", '').strip(),
    }

def precedent_doc_content_formatting(text: str, chunk_size: int, chunk_overlap: int):
    matches = re.findall(r"【(.+?)】(.*?)(?=【.+?】|$)", text)

    formatted_text = dict()
    for m in matches:
        key = str.replace(m[0], " ", "").strip()
        value = chunking(m[1], chunk_size, chunk_overlap)
        formatted_text[key] = value

    return formatted_text


def chunking(text: str, chunk_size: int, chunk_overlap: int):
    base = text.split("<br/>")
    chunks = []
    for s in base:
        s = s.strip()
        if len(s) == 0:
            continue

        section_title = (
            r"^((([0-9]{1,2}|[가-하])\.)|(\(([0-9]{1,2}|[가-하])\)))\s"  # 1. 가. (1)
        )
        # [\u2460-\u2473] : 1-20 ①②③④⑤⑥⑦⑧⑨⑩⑪⑫⑬⑭⑮⑯⑰⑱⑲⑳
        # [\u3251-\u32BF] : 21-50
        # [\u326e-\u327b] ㉮-㉻
        enumeration = r"[\u2460-\u2473\u326e-\u327b]\s"

        # prefix 제거
        s = re.sub(section_title, "", s)
        # 서수 기호 제거
        s = re.sub(enumeration, "", s)

        text_spliter = RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap=chunk_overlap,
            length_function=len,
            # is_separator_regex=True,
            separators=[","],
            keep_separator=False,
        )
        chunks += (text_spliter.split_text(s))
    
    for i, c in enumerate(chunks):
        chunks[i] = c.strip()

    return chunks

def precedent_doc_formatting(content : dict[str, str]) -> dict[str, str|dict[str, list]]:
    content = precedent_doc_trim_content(content)
    content['판례내용'] = precedent_doc_content_formatting(content['판례내용'], 80, 20)
    return content

def extract_sentences_from_format(format : dict[str, str|dict[str, list]] ) -> list[str]:
    sentences = []
    for f in format.values():
        if len(f) == 0: continue # empty string
        if isinstance(f, dict):
            for v in f.values():
                if len(v) == 0 : continue # empty list
                for e in v:
                    sentences.append(e)
        else:
            sentences.append(f)

    return sentences
```
